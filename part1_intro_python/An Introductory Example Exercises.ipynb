## My Solution to Exercise 1

```python
def factorial(n):
    if n==0:
        return 1
    else:
        return n*factorial(n-1)

factorial(4)
```

## Text Solution to Exercise 1

```python
def factorial(n):
    k = 1
    for i in range(n):
        k = k * (i + 1)
    return k

factorial(4)
```

## My Solution to Exercise 2

```python
from numpy.random import uniform

def binomial_rv(n, p):
    count = 0
    for i in range(n):
        if uniform() < p:
            count+=1
    return count

binomial_rv(10,.5)
```

## Text Solution to Exercise 2

```python
from numpy.random import uniform

def binomial_rv(n, p):
    count = 0
    for i in range(n):
        U = uniform()
        if U < p:
            count = count + 1    # Or count += 1
    return count

binomial_rv(10, 0.5)
```

## My Solution to Exercise 3

```python
import numpy as np

def inside_arc(a,b):
    if np.sqrt(a**2+b**2) < 1:
        return True
    else:
        return False

monte_values = np.array([])
simulations = 100000

for i in range(simulations):
    x,y = np.random.uniform(),np.random.uniform()
    monte_values = np.append(monte_values, inside_arc(x,y))
area = monte_values.sum()/simulations*4
pi = area
pi
```

## Text Solution to Exercise 3

```python
import numpy as np

n = 100000

count = 0
for i in range(n):
    u, v = np.random.uniform(), np.random.uniform()
    d = np.sqrt((u - 0.5)**2 + (v - 0.5)**2)
    if d < 0.5:
        count += 1

area_estimate = count / n

print(area_estimate * 4)  # dividing by radius**2
```

## My Solution to Exercise 4

```python
from numpy.random import uniform

record = []
for i in range(10):
    if uniform() < .5:
        record.append(False)
    else:
        record.append(True)

streak = 0
payout = 0
for rec in record:
    if rec:
        streak += 1
        if streak >= 3:
            payout = 1
            break
    else:
        streak = 0
    
print(record)
print(payout)
```

## Text Solution to Exercise 4

```python
from numpy.random import uniform

payoff = 0
count = 0

for i in range(10):
    U = uniform()
    count = count + 1 if U < 0.5 else 0
    if count == 3:
        payoff = 1

print(payoff)
```

## My Solution to Exercise 5

```python
import numpy as np
import matplotlib.pyplot as plt

T = 200
alpha = 0.9
x = 0
vec = np.array([0])

for i in range(T):
    x = alpha*x + np.random.normal()
    vec = np.append(vec, x)
    
plt.plot(vec)
plt.show()
```

## Text Solution to Exercise 5

```python
import matplotlib.pyplot as plt
import numpy as np

alpha = 0.9
ts_length = 200
current_x = 0

x_values = []
for i in range(ts_length + 1):
    x_values.append(current_x)
    current_x = alpha * current_x + np.random.randn()
plt.plot(x_values, 'b-')
plt.show()
```

## My Solution to Exercise 6

```python
import numpy as np
import matplotlib.pyplot as plt

ts_length = 200

for alpha in [0, 0.8, 0.98]:
    x_values = []
    current_x = 0
    for i in range(ts_length + 1):
        x_values.append(current_x)
        current_x = alpha * current_x + np.random.randn()
    plt.plot(x_values, label="alpha = " + str(alpha))

plt.legend()
plt.show()
```
